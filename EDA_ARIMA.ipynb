## 정상성 확인 (Stationary Time Series)

### ADF 검정

```python
plt.figure(figsize = (24,16))
for i in range(1, len(df.columns)):
    plt.subplot(len(df.columns),1,i)
    plt.plot(np.log(df.iloc[:,i]).diff())
plt.show()
```

```python
# 계절성 제거
plt.figure(figsize = (24,16))
for i in range(1, len(df.columns)):
    plt.subplot(len(df.columns),1,i)
    plt.plot(np.log(df.iloc[:,i]).diff().diff(12).dropna())
plt.show()
```

```python
def adf_test(x):
    stat, p_value, lags, nobs, crit, icb = adfuller(x) #adfuller 함수를 이용하면 6가지 output이 나옵니다
    print(stat)
    print(p_value)
    print(lags)
    print(nobs)
    print(crit)
    print(icb)

```

```python
adf_test(df.iloc[:,1])
```

```python
test = []
for i in range(0, len(df.columns)):
   test.append(adfuller(df.iloc[:,i])[0:2]) #stat, p-value 저장
```

### KPSS 검정

```python
from statsmodels.tsa.stattools import kpss

def kpss_test(df):    
    statistic, p_value, n_lags, critical_values = kpss(df.values)
    
    print(f'KPSS Statistic: {statistic}')
    print(f'p-value: {p_value}')
    print(f'num lags: {n_lags}')
    print('Critial Values:')
    for key, value in critical_values.items():
        print(f'   {key} : {value}')
```

```python
statistic, p_value, n_lags, critical_values = kpss(df.iloc[:,1])
```

```python
test3 = []
for i in range(0, len(df.columns)):
   t = kpss_test(df.iloc[:,i])
   statistic, p_value, n_lags, critical_values = kpss(df.iloc[:,i])
   test3.append((statistic, p_value)) #stat, p-value 저장
```

```python
test3 = pd.DataFrame(test3)
test3[2] = test3[1] > 0.05
test3
```

```python
from statsmodels.tsa.arima_model import ARIMA
from pandas import datetime

# fit model
model = ARIMA(pd.DataFrame(df.iloc[:,1]), order=(2,1,2))
model_fit = model.fit(disp=0)

# predict
start_index = datetime(2004, 1, 31)
end_index = datetime(2020, 2, 29)
forecast = model_fit.predict(start=start_index, end=end_index, typ='levels')

# visualization
plt.figure(figsize=(22,8))
plt.plot(df.iloc[:,1],label = "original")
plt.plot(forecast,label = "predicted")
plt.title("Time Series Forecast")
plt.xlabel("Date")
plt.ylabel("Mean Temperature")
plt.legend()
plt.show()
```

```python
from sklearn import metrics


def scoring(y_true, y_pred):
    r2 = round(metrics.r2_score(y_true, y_pred) * 100, 3)
    #     mae = round(metrics.mean_absolute_error(y_true, y_pred),3)
    corr = round(np.corrcoef(y_true, y_pred)[0, 1], 3)
    mape = round(
        metrics.mean_absolute_percentage_error(y_true, y_pred) * 100, 3)
    rmse = round(metrics.mean_squared_error(y_true, y_pred, squared=False), 3)

    df = pd.DataFrame({
        'R2': r2,
        "Corr": corr,
        "RMSE": rmse,
        "MAPE": mape
    },
                      index=[0])
    return df
```
